```python
import pandas as pd
```

```python
df = pd.read_csv("education_by_state.csv", header=1)

df.columns = [
    'State',
    'Population over the age of 25',
    'HS Diploma Population',
    'HS Diploma Pct',
    'BS Degree Population',
    'BS Degree Pct',
    'Advanced Degree Population',
    'Advanced Degree Pct'
]

df = df.dropna(subset=['State'])

for column in ['HS Diploma Pct', 'BS Degree Pct', 'Advanced Degree Pct']:
    df[column] = df[column].str.replace('%', '').astype(float)

for column in [
    'Population over the age of 25',
    'HS Diploma Population',
    'BS Degree Population',
    'Advanced Degree Population'
]:
    df[column] = df[column].str.replace(',', '').astype(int)

df.set_index('State', inplace=True)
df
```

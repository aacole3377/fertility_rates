```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_fertility_rate"
df = pd.read_html(url)
table1 = df[1]
table2 = df[2]
```

```python
df[1].columns
```

```python
merged = pd.merge(table1, table2, on='State federal district  or territory', how='outer')
merged
```

```python
merged.rename(columns={'State federal district  or territory' : 'State'}, inplace=True)
merged.set_index('State', inplace=True)
```

```python
column_order = [ 
    'TFR 2008[5]', 'TFR 2009[6]', 'TFR 2010[7]', 'TFR 2011[8]', 'TFR 2012[9]', 
    'TFR 2013[10]', 'TFR 2014[11]', 'TFR 2015[12]', 'TFR 2016[13]', 'TFR 2017[14]', 
    'TFR 2018[15]', 'TFR 2019[1]', 'TFR 2020[2]', 'TFR 2021[3]', 'TFR 2022[4]'
]

merged = merged[column_order]
```

```python
# Extract the year from the column names
merged.columns = merged.columns.str.extract(r'(\d{4})', expand=False)
merged
```
